```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo dữ liệu ví dụ
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value': [23, 45, 56, 78]
}

# Chuyển đổi dữ liệu thành DataFrame
df = pd.DataFrame(data)

# Tạo biểu đồ cột
df.plot(kind='bar', x='Category', y='Value', color='skyblue')

# Thêm tiêu đề và nhãn cho các trục
plt.title('Biểu đồ cột mẫu')
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')

# Hiển thị biểu đồ
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dữ liệu mẫu: ma trận tương quan giữa các biến
data = {'Var1': [1, 2, 3, 4], 'Var2': [4, 3, 2, 1], 'Var3': [2, 3, 4, 5]}
df = pd.DataFrame(data)

# Tính ma trận tương quan
corr = df.corr()

# Vẽ biểu đồ Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Thêm tiêu đề
plt.title('Biểu đồ Heatmap')

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Dữ liệu mẫu
names = ['Hoa', 'Nam', 'Hai', 'Linh', 'Nhan', 'Phuong', 'Thu', 'Vu', 'Bao', 'Hieu', 'Lan', 'Duy', 'Tuan', 'Kien', 'Minh', 'Trung', 'Quyen', 'Tung', 'Hien', 'Mai']
genders = ['Nam', 'Nữ']
classes = ['10A1', '10A2', '11B1', '11B2', '12C1', '12C2']

def generate_scores():
    return [random.randint(5, 10) for _ in range(3)] 

data = {
    'Tên': names,
    'Giới tính': [random.choice(genders) for _ in names],
    'Điểm Toán': [score[0] for score in [generate_scores() for _ in names]],
    'Điểm Anh Văn': [score[1] for score in [generate_scores() for _ in names]],
    'Điểm Ngữ Văn': [score[2] for score in [generate_scores() for _ in names]],
    'Lớp': [random.choice(classes) for _ in names]
}

students_df = pd.DataFrame(data)

# Vẽ biểu đồ cột cho các môn học
students_df.set_index('Tên')[['Điểm Toán', 'Điểm Anh Văn', 'Điểm Ngữ Văn']].plot(kind='bar', figsize=(10, 6))

plt.title('Điểm của học sinh trong các môn học')
plt.xlabel('Tên học sinh')
plt.ylabel('Điểm')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
