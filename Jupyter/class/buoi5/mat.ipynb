```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu

df = pd.read_csv("spotify.csv")  


# Chuyn ct "Date" thnh datetime v t lm index
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date")

# in gi tr thiu (c th dng 0 hoc ni suy)
df = df.fillna(0)  # Hoc df = df.interpolate()

# V biu  streams theo thi gian
plt.figure(figsize=(15, 5))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

# Tu chnh biu 
plt.xlabel("Date")
plt.ylabel("Streams")
plt.title("Spotify Song Streams Over Time")
plt.legend()  # Hin th ch thch
plt.xticks(rotation=45)  # Xoay nhn trc X cho d c
plt.grid(True)  # Hin th li
plt.show()
```

```python
# V biu  streams theo thi gian (ch v 1 ct)
plt.figure(figsize=(15, 5))
plt.plot(df.index, df["Shape of You"], label="Shape of You", color="blue")
plt.xlabel("Date")
plt.ylabel("Streams")
plt.title("Shape of You Spotify Streams Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# D liu ngh s v nm sinh
data = {
    "Artist": ["Ed Sheeran", "Luis Fonsi", "The Chainsmokers", "Kendrick Lamar", "French Montana"],
    "Birth Year": [1991, 1978, 1989, 1987, 1984]
}

df = pd.DataFrame(data)

# Tnh tui
current_year = datetime.now().year
df["Age"] = current_year - df["Birth Year"]

# V biu  ct
plt.figure(figsize=(10, 5))
plt.bar(df["Artist"], df["Age"], color=['blue', 'orange', 'green', 'red', 'purple'])

# Thm tiu  v nhn
plt.xlabel("Ngh s")
plt.ylabel("Tui")
plt.title("Biu  tui ca cc ngh s")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hin th biu 
plt.show()

```
