```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu

df = pd.read_csv("spotify.csv")  


# Chuyển cột "Date" thành datetime và đặt làm index
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date")

# Điền giá trị thiếu (có thể dùng 0 hoặc nội suy)
df = df.fillna(0)  # Hoặc df = df.interpolate()

# Vẽ biểu đồ streams theo thời gian
plt.figure(figsize=(15, 5))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

# Tuỳ chỉnh biểu đồ
plt.xlabel("Date")
plt.ylabel("Streams")
plt.title("Spotify Song Streams Over Time")
plt.legend()  # Hiển thị chú thích
plt.xticks(rotation=45)  # Xoay nhãn trục X cho dễ đọc
plt.grid(True)  # Hiển thị lưới
plt.show()
```

```python
# Vẽ biểu đồ streams theo thời gian (chỉ vẽ 1 cột)
plt.figure(figsize=(15, 5))
plt.plot(df.index, df["Shape of You"], label="Shape of You", color="blue")
plt.xlabel("Date")
plt.ylabel("Streams")
plt.title("Shape of You Spotify Streams Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Dữ liệu nghệ sĩ và năm sinh
data = {
    "Artist": ["Ed Sheeran", "Luis Fonsi", "The Chainsmokers", "Kendrick Lamar", "French Montana"],
    "Birth Year": [1991, 1978, 1989, 1987, 1984]
}

df = pd.DataFrame(data)

# Tính tuổi
current_year = datetime.now().year
df["Age"] = current_year - df["Birth Year"]

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 5))
plt.bar(df["Artist"], df["Age"], color=['blue', 'orange', 'green', 'red', 'purple'])

# Thêm tiêu đề và nhãn
plt.xlabel("Nghệ sĩ")
plt.ylabel("Tuổi")
plt.title("Biểu đồ tuổi của các nghệ sĩ")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hiển thị biểu đồ
plt.show()

```
