```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp CSV
file_path = 'hw3_data.csv'  # Đảm bảo tệp nằm trong cùng thư mục hoặc cung cấp đường dẫn đầy đủ
data = pd.read_csv(file_path)

# Câu hỏi 1: Có bao nhiêu người trên 30 tuổi trong dữ liệu?
over_30_count = data[data['Age'] > 30].shape[0]

# Câu hỏi 2: Tổng số tiền lương của tất cả các người trong dữ liệu là bao nhiêu?
total_salary = data['Salary'].sum()

# Câu hỏi 3: Có bao nhiêu người có lương lớn hơn 50,000 và tuổi nhỏ hơn 40?
high_salary_young_count = data[(data['Salary'] > 50000) & (data['Age'] < 40)].shape[0]

# In bảng dữ liệu lên màn hình dưới dạng bảng DOS
print("\nDữ liệu nhân viên:")
print(tabulate(data, headers='keys', tablefmt='grid'))

# Tạo sơ đồ thống kê
plt.figure(figsize=(10, 6))

# Biểu đồ cột: Phân bổ độ tuổi
plt.subplot(1, 2, 1)
data['Age'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black')
plt.title('Phân bổ độ tuổi')
plt.xlabel('Tuổi')
plt.ylabel('Số lượng')

# Biểu đồ tròn: Phân bổ theo phòng ban
plt.subplot(1, 2, 2)
department_counts = data['Department'].value_counts()
department_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgreen', 'lightblue', 'pink'])
plt.title('Phân bổ theo phòng ban')
plt.ylabel('')

plt.tight_layout()

plt.show()

# In kết quả ra file notebook
with open("output_notebook.txt", "w") as file:
    file.write(f"Số người trên 30 tuổi: {over_30_count}\n")
    file.write(f"Tổng số tiền lương: {total_salary}\n")
    file.write(f"Số người có lương lớn hơn 50,000 và tuổi nhỏ hơn 40: {high_salary_young_count}\n")

# In kết quả lên màn hình
print("\nKết quả:")
print("Số người trên 30 tuổi:", over_30_count)
print("Tổng số tiền lương:", total_salary)
print("Số người có lương lớn hơn 50,000 và tuổi nhỏ hơn 40:", high_salary_young_count)
```
